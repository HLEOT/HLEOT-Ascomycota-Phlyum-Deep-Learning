```python
# load the csv data but do not use the first column as index
df = pd.read_csv(csv_file, delimiter=';')

# get the gene_ids from the csv file
gene_ids = df['gene_id'].tolist()

# remove the prefix "gene:" from each gene_id
gene_ids = [gene_id.replace('gene:', '') for gene_id in gene_ids]

# dictionary to store gene ids and their indices
gene_id_indices = {gene_id: index for index, gene_id in enumerate(gene_ids)}

# dictionary to store gene ids for each species
species_genes = {}

# iterate over all .gff files in the directory
for filename in os.listdir(folder2):
    if filename.endswith('.gff3'):
        # create a set to hold unique gene ids
        gene_ids_in_file = set()

        with open(os.path.join(folder2, filename), 'r') as file:
            for line in file:
                # use regex to extract gene id
                match = re.search(r'ID=gene:(.*?);', line)
                if match:
                    gene_ids_in_file.add(match.group(1))

        # store the unique gene ids for the current species
        species_genes[filename.rstrip('.gff3')] = gene_ids_in_file


# dictionary to store the row indices for each species
species_rows = {}

# iterate over the species
for species, genes_in_file in species_genes.items():
    # find the matching genes
    matching_genes = [gene_id for gene_id in gene_ids if gene_id in genes_in_file]

    # find the row indices for the matching genes using the gene_id_indices dictionary
    match_rows = [gene_id_indices[gene_id] for gene_id in matching_genes if gene_id in gene_id_indices]

    # store the row indices for the current species
    species_rows[species] = match_rows

# create dataframe from the dictionary
df_output = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in species_rows.items() ]))

# fill NaN with empty string
df_output = df_output.fillna('')
```

```python
# Ensure the output directory exists
os.makedirs(output_folder, exist_ok=True)

try:
    # Save the DataFrame to a CSV file
    df_output.to_csv(os.path.join(output_folder, 'expression_species_rows.csv'), index=False)
    print(f'Data successfully saved to {os.path.join(output_folder, "expression_species_rows.csv")}')
except Exception as e:
    print(f'An error occurred while saving the data: {e}')
```
